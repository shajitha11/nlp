```python
import pandas as pd
```

```python
tickets = [
    "I can’t login to my Gmail account since yesterday.",
    "My iPhone 12 battery drains too fast.",
    "Payment failed while purchasing on Amazon.",
    "The new Tesla Model 3 software update is buggy.",
    "Can’t reset my password for LinkedIn account.",
    "My Netflix subscription is not working after payment.",
    "I received a wrong item from Flipkart order #12345.",
    "Spotify app crashes when I play music on Android.",
    "My bank account shows incorrect balance after transaction.",
    "Unable to connect to WiFi on my MacBook Pro."
]
```

```python
df=pd.DataFrame({
    'ticket_id':range(1,len(tickets)+1),
    'text':tickets
})

print(df)
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")
```

```python
#part of speech pos

def get_pos_tags(text):
    doc=nlp(text)
    return [(token.text,token.pos_)for token in doc]
df["pos_tags"]=df["text"].apply(get_pos_tags)
df[["ticket_id","text","pos_tags"]]
```

```python
#name entity recognition ner

def get_named_ent(text):
    doc=nlp(text)
    return[(ent.text,ent.label)for ent in doc.ents]
df["named_entity"]=df["text"].apply(get_named_ent)
df[["ticket_id","text","named_entity"]]
```
